```python
coordinates = []
cities = []

lat = np.random.uniform(low=-90, high=90, size=1369)
lng = np.random.uniform(low=-180, high=180, size=1369)
coordinates = zip(lat, lng)

for coord in coordinates:
    city = citipy.nearest_city(coord[0], coord[1]).city_name
    
    if city not in cities:
        cities.append(city)
#ensure there is over 500 cities in our list (check! 588)
len(cities)
```

```python
citiescomplete = []
temps = []
windspeeds = []
humiditys = []
countries =[]
latitudes = []
longitudes = []
cloudiness = []

# Set counters need multiple sets because 50 is max 
count = 0
set_count = 1
for city in cities:
    try:
        data = requests.get(f"http://api.openweathermap.org/data/2.5/weather?units=Imperial&appid="+ api_key +"&q="+city).json()
        time.sleep(2)
        temps.append(data['main']['temp'])
        windspeeds.append(data['wind']['speed'])
        humiditys.append(data['main']['humidity'])
        countries.append(data['sys']['country'])
        latitudes.append(data['coord']['lat'])
        longitudes.append(data['coord']['lon'])
        cloudiness.append(data['clouds']['all'])
        
        if count > 49:
            set_count += 1
            count = 1
            citiescomplete.append(city)
        else:
            count += 1
            citiescomplete.append(city)
            print(f"Inserting datapoint {count} of Set {set_count} | {city}")
    except Exception: 
            print("Data not found.Searching next...")
print("-------Data retrieval Complete----------------")
```

```python
weatherdf = pd.DataFrame({"City": citiescomplete,
                          "Country" : countries,
                          "Temperature" : temps,
                          "Wind Speed" : windspeeds,
                          "Humidity" : humiditys,
                          "Cloudiness" : cloudiness,
                          "Latitude" : latitudes, 
                          "Longitude" : longitudes})
#Total numbers still above 500 (537)
weatherdf.count()
#Output data from dataframe to csv for later use
weatherdf.to_csv("Output/Globalcities.csv", index=False, header=True)
```

```python

#Create northern and southern hemispheres
hemi = weatherdf['Latitude'] > 0
northernhemi = weatherdf[hemi]
southernhemi = weatherdf[~hemi]
southernhemi
```

```python
#Regressions
x_valuesh1 = northernhemi["Latitude"]
y_valuesh1 = northernhemi["Temperature"]

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_valuesh1, y_valuesh1)
regress_values = x_valuesh1 * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
correlation = st.pearsonr(x_valuesh1, y_valuesh1)

plt.scatter(x_valuesh1, y_valuesh1)
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.title("Northern Hemisphere Temperature vs. Latitude (for 10/23/2021)")
plt.plot(x_valuesh1,regress_values,"r-")
plt.annotate(line_eq,(5,0),fontsize=12,color="red")
plt.savefig("Images/NHLatitude_versus_Temperature.png")
```
